```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('House_dat_set_batch_4.csv.csv')
data
```

```python
data.head(10)
```

```python
data.columns
```

```python
#pie chart
mycolors = ["blue", "hotpink", "orange", "#4CAF50"]
plt.pie(data['furnishingstatus'].value_counts(),labels = ['furnished','semi-furnished','unfurnished'],colors=mycolors,autopct='%1.2f%%')
#plt.legend()
plt.legend(title = "chart furnishingstatus:")
plt.grid()
```

```python
mycolors = ["blue", "hotpink", "orange", "#4CAF50",'red','black']
plt.pie(data['bedrooms'].value_counts(),labels = [1,2,3,4,5,6],colors=mycolors,autopct='%1.2f%%')
#plt.legend()
plt.legend(title = "number of bedrooms:")
plt.grid()
```

```python
#box plot
sns.boxplot(data['area'])
plt.title("box plot of area")
plt.grid()
plt.show()
```

```python
sns.boxplot(data['price'])
plt.title("box plot of price")
plt.grid()
plt.show()
```

```python
# bar chart
sns.barplot(x = 'furnishingstatus',
            y = 'price', hue='furnishingstatus',
            data = data)
plt.grid()
plt.show()
```

```python
#scatter plot
sns.scatterplot(x="area",
                    y="price", hue="furnishingstatus",
                    data=data)
plt.title("Scatter plot of furnishingstatus")
plt.grid()
plt.show()
```

```python
sns.lmplot(x="area",y="price",hue="furnishingstatus",data=data,markers=["o","x","+"])
plt.grid()
plt.show()
```

```python
plt.hist(data['area'], density = 1, color ='green', alpha = 0.7)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('matplotlib.pyplot.hist() function Example\n\n',
          fontweight ="bold")
plt.grid()
plt.show()
```

```python
plt.hist(data['price'], density = 1, color ='orange', alpha = 0.7)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
  
plt.title('matplotlib.pyplot.hist() function Example\n\n',
          fontweight ="bold")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(numeric_only= True),annot=True,fmt=".2f")
```

```python
sns.pairplot(data,hue="furnishingstatus")
plt.grid()
plt.show()
```
