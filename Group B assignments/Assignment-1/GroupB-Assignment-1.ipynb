```python
#Creating a subset of dataset
```

```python
DFsubset1 = dataFrame[['Page total likes','Category','Paid','comment','like','share']].iloc[5:50]
```

```python
DFsubset2 = dataFrame[['Page total likes','Category','Paid','comment','like','share']].iloc[50:60]
```

```python
DFsubset2
```

```python
#Merge Data
mergeDF = pd.concat([DFsubset1 , DFsubset2])
```
